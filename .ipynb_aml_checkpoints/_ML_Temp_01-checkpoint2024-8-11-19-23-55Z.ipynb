```python
# Import packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Create a dataframe with 3 columns and 100 rows of dummy data
data = {
    'Column1': np.random.randint(1, 100, 100),
    'Column2': np.random.rand(100),
    'Column3': np.random.choice(['A', 'B', 'C', 'D'], 100)
}

df = pd.DataFrame(data)

# Display the dataframe
df.head()
```

```python
# Build a simple graph

# Group the data by 'Column3' and calculate the mean of 'Column2'
grouped_data = df.groupby('Column3')['Column2'].mean()

# Create a bar plot
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color='skyblue')
plt.title('Mean of Column2 by Column3')
plt.xlabel('Column3')
plt.ylabel('Mean of Column2')
plt.xticks(rotation=0)
plt.show()

```

```python
# Build another simple graph
# This is a silly graph... 

# Sort the dataframe by Column3 to ensure alphabetical order
df = df.sort_values(by='Column3')

# Create a line plot with Column3 as the x-axis
plt.figure(figsize=(10, 6))
for label in df['Column3'].unique():
    subset = df[df['Column3'] == label]
    plt.plot(subset['Column3'], subset['Column1'], label=label)

plt.title('Line Graph of Column1 by Column3')
plt.xlabel('Column3')
plt.ylabel('Column1')
plt.legend(title='Column3')
plt.show()
```
